2.1 - Cleaning only the text needed for this task (emojis and ! kept)

```python
!pip install emoji
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import emoji

nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(text, lang='portuguese'):
    stop_words = set(stopwords.words(lang))
    words = word_tokenize(text)
    filtered_sentence = [w for w in words if not w in stop_words]
    return ' '.join(filtered_sentence)

def cleaned_df(df, field):
    df[field] = df[field].str.lower()
    df[field] = df[field].apply(lambda x: emoji.demojize(x))
    df[field] = df[field].apply(lambda x: re.sub(r"[^\w\s!@#\$%\^&\*\(\)-=_\+;<>,\.\?/'\"]",'', x))
    df[field] = df[field].apply(lambda x: remove_stopwords(x))
    return df
```

```python
# Main Dataset
df['created_at'] = pd.DatetimeIndex(df['created_at'])
cleaned_df(df, "text")

# Aux. Dataset
cleaned_df(dftr, "tweet_text")
```

3.0 - Training the Model

Once the training is done, the fine-tuned model are saved to a specified path

3.1 - Balacing sentiment dataset

```python
samples = 150000
dftr_positivo = dftr[dftr['sentiment'] == "Positivo"].head(samples)
dftr_negativo = dftr[dftr['sentiment'] == "Negativo"].head(samples)

dftr_samples = pd.concat([dftr_positivo, dftr_negativo])
```

```python
Counter(dftr_samples['sentiment'])
```

3.2 - Training & Evaluating

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import cross_val_score
from sklearn import metrics
import joblib

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dftr_samples['tweet_text'], dftr_samples['sentiment'],\
                                                    test_size=0.30, random_state=42)
# Create a pipeline with the necessary steps
classify = Pipeline(
    [('vect', CountVectorizer (ngram_range=(1,1))),
     ('clf', RandomForestClassifier(n_estimators=100, n_jobs=-1)),
     ])

# Fit the pipeline to the training data
classify.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classify.predict(X_test)

# Generate the classification report
print(metrics.classification_report(y_test, y_pred, target_names=['negative', 'positive']))

# Save the trained model to a file
joblib.dump(classify, '/content/drive/MyDrive/1.WAGSLIFE/Meus Estudos/MECD/MECD11_ADC/ADC/tweets/pesos/classify.joblib')
```
